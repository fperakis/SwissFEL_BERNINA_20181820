# Loads the processed h5 file and plots data

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import h5py
import time
import sys
```

```python
run = '0000_test01'
h5path = '/sf/bernina/data/p17743/scratch/hdf5/run%s.h5'%run
h5file = h5py.File(h5path,'r')
```

```python
img2d = h5file['JF7/2D_img']
q  = h5file['JF7/Q_bins']
Iq = h5file['JF7/I_Q'] 
```

```python
plt.figure()
plt.title('Average 2D image')
plt.imshow(img2d,clim=[0,10],origin='lower')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=[7,7])

plt.subplot(2,1,1)
plt.imshow(Iq,aspect='auto')
plt.title('Angular integration')
plt.xlabel('Q bin')
plt.ylabel('shot number')
plt.colorbar()

plt.subplot(2,1,2)
plt.title('Average')
plt.plot(np.average(Iq,axis=0))
plt.tight_layout()
plt.show()
```
